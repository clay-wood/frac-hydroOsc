```python
import scipy.io as sio

path1 = ['/home/clay/Documents/research/fracture_doe/p4966/v6_ns_res/run','_ns_res.mat'];
path2 = ['/home/clay/Documents/research/fracture_doe/v7_res/run','_res_v7.mat'];

def datloader(path,exp_num,dat_name):
   
    fullpath = path[0]+exp_num+path[1];
    temp=sio.whosmat(fullpath,squeeze_me=True)

    new=[]
    for ii in range(len(temp)):
        new.append(temp[ii][0])

    import re
    dat_match = [string for string in new if re.match(re.compile(dat_name), string)][0]

    return sio.loadmat(fullpath,squeeze_me=True)[dat_match]

time4 = np.array(datloader(path2,'4','time'))
ns4 = np.array(datloader(path1,'4','NS'))
pp4 = np.array(datloader(path1,'4','PPa'))
ts4 = np.array(datloader(path1,'4','TS'))
```

```python
xcor = np.correlate(((acvec[165:450,9860]-np.mean(acvec[165:450,9860]))/(np.std(acvec[165:450,9860])*len(acvec[165:450,9860]))),
                    ((acvec[165:450,9895]-np.mean(acvec[165:450,9895]))/(np.std(acvec[165:450,9895]))),'same');

maxxcor = np.where(xcor == np.max(xcor))[0][0]

xcortime = np.arange(-maxxcor,285-maxxcor,1)/25

# ------------------------------------------------------------------------------------------------------------------------------

ts = 0.04

NN = (len(xcor)+1)/2;

tps = np.arange(-NN*ts,(NN+1)*ts,ts);

[x1, x2, x3] = [np.argmax(xcor)-1, np.argmax(xcor), np.argmax(xcor)+1]

[y1 ,y2, y3] = xcor[[x1, x2, x3]]

b = ((y1 - y3)*(x1**2 - x2**2) - (y1 - y2)*(x1**2 - x3**2))/((x1 - x3)*(x1**2 - x2**2) - (x1 - x2)*(x1**2 - x3**2));

a = (y1 - y3 - b*(x1 - x3))/(x1**2 - x3**2);

c = y1 - a*x1**2 - b*x1;

ind_max_corr = -b/(2*a);
max_interpolation = a*ind_max_corr**2 +b*ind_max_corr + c;

timedelay = ts*(ind_max_corr - NN);

interpolationX = np.arange(x1,x3+1,0.01);

interpolationY = a*interpolationX**2 + b*interpolationX + c;

interpolationt = ts*(interpolationX - NN+1);
```

```python
# %matplotlib inline

fig, ax1 = plt.subplots(1,1,figsize=(20,10)) #,dpi=200)
mpl.rcParams.update({'font.size': 40})
for spine in ['left','right','top','bottom']:
    ax1.spines[spine].set_color('k')
    ax1.spines[spine].set_linewidth(1.5)

ax1.plot(xcortime,xcor, c='k', lw=4)
ax1.scatter(interpolationt[np.argmax(interpolationY)],np.amax(interpolationY),s=800,c='b',marker='*',zorder=3)
ax1.set_xlim([-1,1])
ax1.set_ylim([-0.5,1.05])

ax1.tick_params(top=True, bottom=True, left=True, right=True, direction='in', axis='both', which='major',pad=15,width=2, length=8)
ax1.minorticks_on();
ax1.tick_params(top=True, bottom=True, left=True, right=True, direction='in', axis='both', which='minor',pad=15,width=1.25, length=5)

# ax1.ticklabel_format(axis='y',style='sci',useOffset=False, )
plt.xticks(np.arange(-1, 1.5, 0.5))
plt.yticks(np.arange(-0.5, 1.25, 0.5))

# AXES LABELS & COLORS   
ax1.set_xlabel('Time Shift ($\mu s$)',color='k',fontsize=40,labelpad=25)
ax1.set_ylabel('Normalized Cross-Correlation',color='k',fontsize=40,labelpad=25)

plt.show()
fig.savefig('/home/clay/Documents/research/fracture_doe/doe-bes17/manuscript/plots/Norm_xcorr.svg', transparent=True)

# ------------------------------------------------------------------------------------------------------------------------------
```
